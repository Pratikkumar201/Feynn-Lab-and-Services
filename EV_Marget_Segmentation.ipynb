```python
from sklearn.cluster import KMeans
clustering_kmeans = KMeans(n_clusters=3)
d['clusters'] = clustering_kmeans.fit_predict(d)

d1=d[d.clusters==1]
d0=d[d.clusters==0]
d2=d[d.clusters==2]

plt.scatter(d1.Price, d1['Total Salary'],color='green')
plt.scatter(d0.Price, d0['Total Salary'],color='red')
plt.scatter(d2.Price, d2['Total Salary'], color='blue')

plt.xlabel('Price')
plt.ylabel('Total Salary')
```

```python
clustering_kmeans.cluster_centers_
```

```python
k_range = range(1,10)
# Sum of sqaured error

sse=[]
for k in k_range:
  km = KMeans(n_clusters=k)
  km.fit(d)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('The Elbow Point Graph')
plt.xlabel('K')
plt.ylabel('sse')
plt.plot(k_range, sse)
```

```python
# We can take elbow as 3
```

```python
def plotseg(clus,Y, labels):
  plt.scatter(clus[Y==0,0], clus[Y==0,1], s=10, c='blue', label='Cluster 1')
  plt.scatter(clus[Y==1,0], clus[Y==1,1], s=10, c='red', label='Cluster 2')
  plt.scatter(clus[Y==2,0], clus[Y==2,1], s=10, c='black', label='Cluster 3')

  plt.title('Customer_Segments')
  plt.xlabel(labels[0])
  plt.ylabel(labels[1])
  plt.show()
```

```python
clus = d.loc[:,["Age","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["Age","Price"])
```

```python
# Model Deployment
# K Means Clustering
```

```python
X=df.iloc[:,df.columns!='Make']
X.head()
```

```python

encoding={"Profession":{"Salaried":0,"Business":1},
          "Marrital Status":{"Single":0,"Married":1},
          "Education":{"Graduate":0,"Post Graduate":1},
          "Personal loan":{"No":0,"Yes":1},
          "House Loan":{"No":0,"Yes":1},
          "Wife Working":{"No":0,"Yes":1}}
```

```python

obj_df=X.replace(encoding)
obj_df.head()
```

```python
# K Means Algorithm
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X_scaled = StandardScaler().fit_transform(obj_df)
X_scaled = pd.DataFrame(X_scaled,columns=['Age', 'Profession', 'Marrital Status', 'Education', 'No of Dependents',
                                          'Personal loan', 'House Loan', 'Wife Working', 'Salary', 'Wife Salary',
                                          'Total Salary','Price'])
x=X_scaled.to_numpy()
X_scaled
```

```python
wc=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=10,random_state=0)
  kmeans.fit(X_scaled)
  wc.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wc)
plt.title('Elbow Method Plot')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Square') # Within cluster sum of squares
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                    max_iter = 300, n_init = 10,random_state = 42)
kmeans.fit(X_scaled)
```

```python
y = kmeans.predict(X_scaled)
y_df = pd.DataFrame(y,columns=['Class'])
```
